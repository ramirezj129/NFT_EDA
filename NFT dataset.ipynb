### Dropping unnecessary columns and na values

```python
nft.drop(['Category', 'Website','Logo'], axis = 1, inplace=True)
```

```python
nft.dropna(inplace =True)
```

```python
nft['Market_Cap_USD'] = nft['Market_Cap_USD']/nft['Market_Cap_USD'].max()
nft['Volume_USD'] = nft['Volume_USD']/nft['Volume_USD'].max()
```

```python
nft
```

```python
new_index = nft.Sales.sort_values(ascending = False).index.values
sorted_nft = nft.reindex(new_index)
sorted_nft.index = range(0,540)
top_nft = sorted_nft.loc[0:9]
top_nft
```

### Displaying top 10 NFTs that had the highest Sales

```python
plt.plot(top_nft["Name"], top_nft["Sales"])
plt.xticks(top_nft['Name'], rotation = 'vertical')
plt.ylabel('Number of Sales')
plt.xlabel('Top 10 NFTs By Name')
plt.show()
```

```python
new_nft = nft.sort_values(by=['Sales','Floor_Price_USD'],ascending = False)
new_nft
```

### Scatter plot of floor price and amount of sales in comparison to that price

```python
new_nft.plot(kind = 'scatter', y = 'Sales', x = 'Floor_Price_USD')
plt.show()
```

```python
import seaborn as sns
new_nft.corr()
```

```python
new_nft_corr = new_nft[['Sales','Volume_USD', 'Market_Cap_USD',
                        'Floor_Price_USD','Average_Price_USD','Owners']].dropna().corr()
new_nft_corr
```

### Heat map to check correlation matrix

```python
sns.heatmap(new_nft_corr, annot = True)
```

### Group nft data set by name with the highest volume average


```python
ax = nft.groupby('Name')['Volume'].agg(['mean', 'count']).query('mean > 0').sort_values('mean', ascending = False).head(10)
ax1 = ax.plot(kind= 'barh', figsize = (12,5), title = 'Average Volume of NFT by Name')
ax1.set_xlabel('Average Volume')
plt.show()

```
